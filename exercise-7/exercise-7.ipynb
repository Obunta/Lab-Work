```python
# reading in the file and make dictionaries of DBH at each census

in_file = open("/home/eeb177-student/Desktop/eeb-177/class-assignments/classwork-21/cocoli.txt")
sizes_94 = {} # empty dictionary
sizes_97 = {}
sizes_98 = {}

# get rid of the header row
in_file.readline()

for line in in_file:
    items = line.split() # each line represents individuals
    # retrieve relevant columns
    tree_id = items[0]
    s94 = float(items[4])
    s97 = float(items[5])
    s98 = float(items[6])
    
    # populate our dictionaries
    sizes_94[tree_id] = s94
    sizes_97[tree_id] = s97
    sizes_98[tree_id] = s98
    
in_file.close()

# converted str to float
```

```python
import matplotlib.pyplot as plt

#?plt

# let's make a histogram of DBHs in 1994
# matplotlib works on lists, so let's make one
sizes_in_1994 = list(sizes_94.values())
```

```python
plt.hist(sizes_in_1994, log = True)
plt.show()
```
